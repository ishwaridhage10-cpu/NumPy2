```python
import numpy as np

# Generate random marks for 10 students and 5 subjects
marks = np.random.randint(30, 101, (10, 5))

# Print the marks array
print("Student Marks:")
print(marks)

# Calculate total and average marks of each student
total = np.sum(marks, axis=1)
average = np.mean(marks, axis=1)

print("\nTotal Marks:")
print(total)

print("\nAverage Marks:")
print(average)

# Find student with highest and lowest total marks
highest = np.argmax(total)
lowest = np.argmin(total)

print("\nStudent with Highest Total:", highest + 1)
print("Highest Total Marks:", total[highest])

print("\nStudent with Lowest Total:", lowest + 1)
print("Lowest Total Marks:", total[lowest])

# Calculate overall class mean and standard deviation
print("\nClass Mean:", np.mean(marks))
print("Class Standard Deviation:", np.std(marks))

# Reshape the array (10x5 to 5x10)
reshaped = marks.reshape(5, 10)

print("\nReshaped Array (5x10):")
print(reshaped)

# Extract marks of top 3 students
top3 = np.argsort(total)[-3:]

print("\nTop 3 Students (Marks):")
print(marks[top3])
```
