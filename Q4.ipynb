```python
import numpy as np

arr = np.arange(1, 25)

print("Original Array:")
print(arr)
print("Shape:", arr.shape)

arr1 = arr.reshape(4, 6)
print("\nShape (4, 6):")
print(arr1)
print("Shape:", arr1.shape)

arr2 = arr.reshape(3, 8)
print("\nShape (3, 8):")
print(arr2)
print("Shape:", arr2.shape)

arr3 = arr.reshape(2, 3, 4)
print("\nShape (2, 3, 4):")
print(arr3)
print("Shape:", arr3.shape)
```
