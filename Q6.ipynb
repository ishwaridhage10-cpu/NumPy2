```python
import numpy as np

arr = np.random.randint(1, 101, (5, 5))

print("Original Matrix:")
print(arr)

print("\nDiagonal Elements:")
print(np.diag(arr))

print("\nElements Greater Than 50:")
print(arr[arr > 50])

arr[arr < 30] = 0

print("\nModified Matrix:")
print(arr)
```
